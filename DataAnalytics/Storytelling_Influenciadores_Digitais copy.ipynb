```python
# tratamento do arquivo thematics.json

thematics = json.load(open("thematics.json", "r", encoding="utf-8"))
thematics_trated = {}
sigla_map = {"ig": "Instagram", "yt": "YouTube", "tt": "TikTok"}
for key, value in thematics.items():
    thematics_trated[sigla_map[key]] = {}
    for key2, value2 in value.items():
        thematics_trated[sigla_map[key]].update({key2: value2["text"]})
thematics_trated
with open("thematics_trated.json", "w", encoding="utf-8") as f:
    json.dump(thematics_trated, f, indent=4, ensure_ascii=False)
```

```python
import pandas as pd
import json
import os


def map_thematics(platform, thematics_ids):
    if not thematics_ids:
        return None
    return [
        thematics_trated[platform].get(str(t_id), "Unknown") for t_id in thematics_ids
    ]


def process_instagram_file(filepath):
    with open(filepath) as f:
        data = json.load(f)
    thematics_descriptions = map_thematics(
        "Instagram", data["features"]["blogger_thematics"]["data"]
    )
    return {
        "platform": "Instagram",
        "social_id": data["basic"]["social_id"],
        "username": data["basic"]["username"],
        "fullname": data["basic"]["fullname"],
        "avatar_url": data["basic"]["avatar_url"],
        "is_verified": data["basic"]["is_verified"],
        "subscribers_count": data["metrics"]["subscribers_count"]["value"],
        "engagement_avg": data["metrics"]["engagement_avg"]["value"],
        "authentic_engagement": data["metrics"]["authentic_engagement"]["value"],
        "audience_geo_country_1_code": data["features"]["audience_geo_country_1_code"],
        "audience_geo_country_1_prc": data["features"]["audience_geo_country_1_prc"],
        "blogger_thematics": thematics_descriptions,
        "video_views_avg": None,
        "likes_avg": None,
        "comments_avg": None,
        "shares_avg": None,
        "description": None,
        "tier": None,
        "hs_value": None,
        "aqs": None,
        "additional_countries": None,
    }


def process_tiktok_file(filepath):
    with open(filepath) as f:
        data = json.load(f)
    return {
        "platform": "TikTok",
        "social_id": data["basic"]["social_id"],
        "username": data["basic"]["username"],
        "fullname": data["basic"]["fullname"],
        "avatar_url": data["basic"]["avatar_url"],
        "is_verified": data["basic"]["is_verified"],
        "subscribers_count": data["metrics"]["subscribers_count"]["value"],
        "engagement_avg": data["metrics"]["er"]["performance"]["30d"]["value"],
        "video_views_avg": data["metrics"]["views_avg"]["performance"]["30d"]["value"],
        "likes_avg": data["metrics"]["alikes_avg"]["performance"]["30d"]["value"],
        "comments_avg": data["metrics"]["comments_avg"]["performance"]["30d"]["value"],
        "shares_avg": data["metrics"]["shares_avg"]["performance"]["30d"]["value"],
        "description": data["basic"]["description"],
        "tier": data["basic"]["tier"],
        "hs_value": data["metrics"]["hs_value"]["value"],
        "aqs": data["features"]["aqs"]["data"]["value"],
        "audience_geo_country_1_code": data["features"]["blogger_geo"]["data"][
            "country"
        ],
        "audience_geo_country_1_prc": None,  # Valor n√£o presente no arquivo TikTok
        "blogger_thematics": None,  # Valor n√£o presente no arquivo TikTok
        "additional_countries": None,
    }


def process_youtube_file(filepath):
    with open(filepath) as f:
        data = json.load(f)
    thematics_descriptions = map_thematics(
        "YouTube", data["features"]["blogger_thematics"]["data"]
    )
    additional_countries = [
        {"country": c["id"], "prc": c["prc"]}
        for c in data["features"]["audience_geo"]["data"]["countries"]
    ]
    return {
        "platform": "YouTube",
        "social_id": data["basic"]["social_id"],
        "username": data["basic"]["username"],
        "fullname": data["basic"]["fullname"],
        "avatar_url": data["basic"]["avatar_url"],
        "is_verified": data["basic"]["is_verified"],
        "subscribers_count": data["metrics"]["subscribers_count"]["value"],
        "engagement_avg": data["metrics"]["engagement_avg"]["performance"]["30d"][
            "value"
        ],
        "video_views_avg": data["metrics"]["video_views_avg"]["performance"]["30d"][
            "value"
        ],
        "likes_avg": data["metrics"]["likes_avg"]["performance"]["30d"]["value"],
        "comments_avg": data["metrics"]["comments_avg"]["performance"]["30d"]["value"],
        "shares_avg": None,  # Valor n√£o presente no arquivo YouTube
        "audience_geo_country_1_code": (
            data["features"]["audience_geo"]["data"]["countries"][0]["id"]
            if data["features"]["audience_geo"]["data"]["countries"]
            else None
        ),
        "audience_geo_country_1_prc": (
            data["features"]["audience_geo"]["data"]["countries"][0]["prc"]
            if data["features"]["audience_geo"]["data"]["countries"]
            else None
        ),
        "blogger_thematics": thematics_descriptions,
        "description": None,
        "tier": None,
        "hs_value": None,
        "aqs": None,
        "additional_countries": additional_countries,
    }


def process_files_list(files_list, process_function):
    data_list = []
    for filepath in files_list:
        data = process_function(filepath)
        data_list.append(data)
    return pd.DataFrame(data_list)


def process_all_files(all_files):
    df_instagram = process_files_list(all_files["instagram"], process_instagram_file)
    df_tiktok = process_files_list(all_files["tiktok"], process_tiktok_file)
    df_youtube = process_files_list(all_files["youtube"], process_youtube_file)
    # Combinar todos os DataFrames
    df_combined = pd.concat(
        [df_instagram, df_tiktok, df_youtube],
        ignore_index=True,
        sort=False,
    )
    return df_combined


df_combined = process_all_files(all_files)
# Exibir 5 registros aleat√≥rios

df_combined_sample = df_combined.sample(5)
# Garantir que haja ao menos um de cada plataforma
count = 0
while not all(
    [
        "Instagram" in df_combined_sample["platform"].values,
        "TikTok" in df_combined_sample["platform"].values,
        "YouTube" in df_combined_sample["platform"].values,
    ]
):
    df_combined_sample = df_combined.sample(5)
    count += 1
    if count > 100:
        raise Exception("N√£o foi poss√≠vel encontrar um exemplo de cada plataforma.")

display(df_combined_sample)
df_combined.to_json("dados_processados.json", index=False, indent=4)
```

```python
# Verificando dados faltantes e nulos no DataFrame combinado
missing_data_summary = df_combined.isnull().sum()
missing_data_percentage = (df_combined.isnull().sum() / len(df_combined)) * 100

# Combinando resumo e percentual de dados faltantes
missing_data_df = pd.DataFrame(
    {
        "Missing Count": missing_data_summary,
        "Missing Percentage": missing_data_percentage,
    }
)


display(missing_data_df)
```

```python
# Criando uma matriz de dados faltantes
missing_matrix = df_combined.isnull()
# Calculando a correla√ß√£o entre os dados faltantes e a plataforma
missing_matrix_copy = missing_matrix.copy()
missing_matrix_copy["platform"] = df_combined["platform"]
missing_correlation_copy = missing_matrix.groupby("platform").mean().transpose()

# Exibindo as correla√ß√µes fortes (acima de 0.7)
strong_correlations = missing_correlation_copy[
    (missing_correlation_copy > 0.7) | (missing_correlation_copy < -0.7)
]
strong_correlations = strong_correlations.dropna(how="all", axis=0).dropna(
    how="all", axis=1
)


strong_correlations
```

```python
# Criando o gr√°fico de calor para verificar a correla√ß√£o entre dados faltantes e a plataforma
plt.figure(figsize=(12, 8))
sns.heatmap(
    missing_matrix.groupby(df_combined["platform"]).mean(),
    cmap="viridis",
    cbar=True,
    yticklabels=True,
)
plt.title("Correla√ß√£o entre Dados Faltantes e a Plataforma")
plt.xlabel("Colunas")
plt.ylabel("Plataforma")
plt.show()
```
