```python
# tratamento do arquivo thematics.json

thematics = json.load(open("thematics.json", "r", encoding="utf-8"))
thematics_trated = {}
sigla_map = {"ig": "Instagram", "yt": "YouTube", "tt": "TikTok"}
for key, value in thematics.items():
    thematics_trated[sigla_map[key]] = {}
    for key2, value2 in value.items():
        thematics_trated[sigla_map[key]].update({key2: value2["text"]})
thematics_trated
with open("thematics_trated.json", "w", encoding="utf-8") as f:
    json.dump(thematics_trated, f, indent=4, ensure_ascii=False)
```

```python
import pandas as pd
import json
import os


def map_thematics(platform, thematics_ids):
    if not thematics_ids:
        return None
    return [
        thematics_trated[platform].get(str(t_id), "Unknown") for t_id in thematics_ids
    ]


def process_instagram_file(filepath):
    with open(filepath) as f:
        data = json.load(f)
    thematics_descriptions = map_thematics(
        "Instagram", data["features"]["blogger_thematics"]["data"]
    )
    return {
        "platform": "Instagram",
        "social_id": data["basic"]["social_id"],
        "username": data["basic"]["username"],
        "fullname": data["basic"]["fullname"],
        "avatar_url": data["basic"]["avatar_url"],
        "is_verified": data["basic"]["is_verified"],
        "subscribers_count": data["metrics"]["subscribers_count"]["value"],
        "engagement_avg": data["metrics"]["engagement_avg"]["value"],
        "authentic_engagement": data["metrics"]["authentic_engagement"]["value"],
        "audience_geo_country_1_code": data["features"]["audience_geo_country_1_code"],
        "audience_geo_country_1_prc": data["features"]["audience_geo_country_1_prc"],
        "blogger_thematics": thematics_descriptions,
        "video_views_avg": None,
        "likes_avg": None,
        "comments_avg": None,
        "shares_avg": None,
        "description": None,
        "tier": None,
        "hs_value": None,
        "aqs": None,
        "additional_countries": None,
    }


def process_tiktok_file(filepath):
    with open(filepath) as f:
        data = json.load(f)
    return {
        "platform": "TikTok",
        "social_id": data["basic"]["social_id"],
        "username": data["basic"]["username"],
        "fullname": data["basic"]["fullname"],
        "avatar_url": data["basic"]["avatar_url"],
        "is_verified": data["basic"]["is_verified"],
        "subscribers_count": data["metrics"]["subscribers_count"]["value"],
        "engagement_avg": data["metrics"]["er"]["performance"]["30d"]["value"],
        "video_views_avg": data["metrics"]["views_avg"]["performance"]["30d"]["value"],
        "likes_avg": data["metrics"]["alikes_avg"]["performance"]["30d"]["value"],
        "comments_avg": data["metrics"]["comments_avg"]["performance"]["30d"]["value"],
        "shares_avg": data["metrics"]["shares_avg"]["performance"]["30d"]["value"],
        "description": data["basic"]["description"],
        "tier": data["basic"]["tier"],
        "hs_value": data["metrics"]["hs_value"]["value"],
        "aqs": data["features"]["aqs"]["data"]["value"],
        "audience_geo_country_1_code": data["features"]["blogger_geo"]["data"][
            "country"
        ],
        "audience_geo_country_1_prc": None,  # Valor nÃ£o presente no arquivo TikTok
        "blogger_thematics": None,  # Valor nÃ£o presente no arquivo TikTok
        "additional_countries": None,
    }


def process_youtube_file(filepath):
    with open(filepath) as f:
        data = json.load(f)
    thematics_descriptions = map_thematics(
        "YouTube", data["features"]["blogger_thematics"]["data"]
    )
    additional_countries = [
        {"country": c["id"], "prc": c["prc"]}
        for c in data["features"]["audience_geo"]["data"]["countries"]
    ]
    return {
        "platform": "YouTube",
        "social_id": data["basic"]["social_id"],
        "username": data["basic"]["username"],
        "fullname": data["basic"]["fullname"],
        "avatar_url": data["basic"]["avatar_url"],
        "is_verified": data["basic"]["is_verified"],
        "subscribers_count": data["metrics"]["subscribers_count"]["value"],
        "engagement_avg": data["metrics"]["engagement_avg"]["performance"]["30d"][
            "value"
        ],
        "video_views_avg": data["metrics"]["video_views_avg"]["performance"]["30d"][
            "value"
        ],
        "likes_avg": data["metrics"]["likes_avg"]["performance"]["30d"]["value"],
        "comments_avg": data["metrics"]["comments_avg"]["performance"]["30d"]["value"],
        "shares_avg": None,  # Valor nÃ£o presente no arquivo YouTube
        "audience_geo_country_1_code": (
            data["features"]["audience_geo"]["data"]["countries"][0]["id"]
            if data["features"]["audience_geo"]["data"]["countries"]
            else None
        ),
        "audience_geo_country_1_prc": (
            data["features"]["audience_geo"]["data"]["countries"][0]["prc"]
            if data["features"]["audience_geo"]["data"]["countries"]
            else None
        ),
        "blogger_thematics": thematics_descriptions,
        "description": None,
        "tier": None,
        "hs_value": None,
        "aqs": None,
        "additional_countries": additional_countries,
    }


def process_files_list(files_list, process_function):
    data_list = []
    for filepath in files_list:
        data = process_function(filepath)
        data_list.append(data)
    return pd.DataFrame(data_list)


def process_all_files(all_files):
    df_instagram = process_files_list(all_files["Instagram"], process_instagram_file)
    df_tiktok = process_files_list(all_files["TikTok"], process_tiktok_file)
    df_youtube = process_files_list(all_files["YouTube"], process_youtube_file)
    # Combinar todos os DataFrames
    df_combined = pd.concat(
        [df_instagram, df_tiktok, df_youtube],
        ignore_index=True,
        sort=False,
    )
    return df_combined


df_combined = process_all_files(all_files)
# Exibir 5 registros aleatÃ³rios

df_combined_sample = df_combined.sample(5)
# Garantir que haja ao menos um de cada plataforma
count = 0
while not all(
    [
        "Instagram" in df_combined_sample["platform"].values,
        "TikTok" in df_combined_sample["platform"].values,
        "YouTube" in df_combined_sample["platform"].values,
    ]
):
    df_combined_sample = df_combined.sample(5)
    count += 1
    if count > 100:
        raise Exception("NÃ£o foi possÃ­vel encontrar um exemplo de cada plataforma.")

display(df_combined_sample)
df_combined.to_feather("dados_processados.feather")
```

```python
# Criando uma matriz de dados faltantes
missing_matrix = df_combined.isnull()
# Calculando a correlaÃ§Ã£o entre os dados faltantes e a plataforma
missing_matrix_copy = missing_matrix.copy()
missing_matrix_copy["platform"] = df_combined["platform"]
missing_correlation = missing_matrix_copy.groupby("platform").mean().transpose()

# Exibindo as correlaÃ§Ãµes fortes (acima de 0.7)
strong_correlations = missing_correlation[
    (missing_correlation > 0.7) | (missing_correlation < -0.7)
]
strong_correlations = strong_correlations.dropna(how="all", axis=0).dropna(
    how="all", axis=1
)


strong_correlations
```

```python
# Criando o grÃ¡fico de calor para verificar a correlaÃ§Ã£o entre dados faltantes e a plataforma
plt.figure(figsize=(12, 8))
sns.heatmap(
    missing_matrix.groupby(df_combined["platform"]).mean(),
    cmap="viridis",
    cbar=True,
    yticklabels=True,
)
plt.title("CorrelaÃ§Ã£o entre Dados Faltantes e a Plataforma")
plt.xlabel("Colunas")
plt.ylabel("Plataforma")
plt.show()
```

```python
# Separando o DataFrame combinado em trÃªs datasets distintos por plataforma
df_instagram = df_combined[df_combined["platform"] == "Instagram"].copy()
df_tiktok = df_combined[df_combined["platform"] == "TikTok"].copy()
df_youtube = df_combined[df_combined["platform"] == "YouTube"].copy()


# FunÃ§Ã£o para dropar colunas com 100% de correlaÃ§Ã£o de dados faltantes com a plataforma
def drop_full_correlated_columns(df, platform_name, correlations):
    columns_to_drop = correlations[platform_name].dropna().index.tolist()
    return df.drop(columns=columns_to_drop)


# Dropar colunas com 100% de correlaÃ§Ã£o com a plataforma especÃ­fica
df_instagram_clean = drop_full_correlated_columns(
    df_instagram, "Instagram", strong_correlations
)
df_tiktok_clean = drop_full_correlated_columns(df_tiktok, "TikTok", strong_correlations)
df_youtube_clean = drop_full_correlated_columns(
    df_youtube, "YouTube", strong_correlations
)


# Salvando os datasets limpos para cada plataforma
df_instagram_clean_path = "dados_instagram_limpos.feather"
df_tiktok_clean_path = "dados_tiktok_limpos.feather"
df_youtube_clean_path = "dados_youtube_limpos.feather"

df_instagram_clean.to_feather(df_instagram_clean_path)
df_tiktok_clean.to_feather(df_tiktok_clean_path)
df_youtube_clean.to_feather(df_youtube_clean_path)

# Paths para os arquivos salvos
df_instagram_clean_path, df_tiktok_clean_path, df_youtube_clean_path
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import numpy as np

# Selecionar caracterÃ­sticas relevantes para a clusterizaÃ§Ã£o
features = ["subscribers_count", "engagement_avg", "authentic_engagement"]

# Preparar o DataFrame para clusterizaÃ§Ã£o
df_cluster = df_instagram_clean[features].copy()

# Substituir valores NaN por 0 (ou outro valor apropriado) para a clusterizaÃ§Ã£o
df_cluster.fillna(0, inplace=True)

# Normalizar os dados
scaler = StandardScaler()
df_cluster_scaled = scaler.fit_transform(df_cluster)

# Determinar o nÃºmero Ã³timo de clusters usando o mÃ©todo do cotovelo
sse = []
k_range = range(1, 11)
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df_cluster_scaled)
    sse.append(kmeans.inertia_)

# Calcular a pontuaÃ§Ã£o da silhueta para diferentes nÃºmeros de clusters
silhouette_scores = []
for k in k_range[1:]:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df_cluster_scaled)
    score = silhouette_score(df_cluster_scaled, kmeans.labels_)
    silhouette_scores.append(score)

# Plotar o grÃ¡fico do mÃ©todo do cotovelo e a pontuaÃ§Ã£o da silhueta
plt.figure(figsize=(16, 6))

plt.subplot(1, 2, 1)
plt.plot(k_range, sse, marker="o")
plt.title("MÃ©todo do Cotovelo")
plt.xlabel("NÃºmero de Clusters")
plt.ylabel("SSE")

plt.subplot(1, 2, 2)
plt.plot(k_range[1:], silhouette_scores, marker="o")
plt.title("PontuaÃ§Ã£o da Silhueta")
plt.xlabel("NÃºmero de Clusters")
plt.ylabel("PontuaÃ§Ã£o da Silhueta")

plt.show()

# Escolher o nÃºmero Ã³timo de clusters (a partir dos grÃ¡ficos)
num_clusters = 3  # Escolhido com base nos grÃ¡ficos

# Aplicar K-means com o nÃºmero Ã³timo de clusters
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
df_instagram_clean["cluster"] = kmeans.fit_predict(df_cluster_scaled)

# AnÃ¡lise dos Clusters
cluster_summary = (
    df_instagram_clean.select_dtypes(exclude="object").groupby("cluster").mean()
)

# Verificar a distribuiÃ§Ã£o dos dados faltantes por cluster
missing_by_cluster = (
    df_instagram_clean.select_dtypes(exclude="object")
    .groupby("cluster")["tematica_faltante"]
    .mean()
)

display(cluster_summary)
display(pd.DataFrame(missing_by_cluster))
```
